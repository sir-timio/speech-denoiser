```python
from IPython import display as disp
import torch
import torchaudio
from denoiser import pretrained
from denoiser.dsp import convert_audio
import librosa
```

```python
import librosa
```

```python
fname = 'files/224141cd-7113-4638-b038-f4d93ff040fc.wav'
disp.Audio(fname)
```

```python
wav, sr = librosa.load(fname)
```

```python
model = pretrained.dns64().cpu()
```

```python
wav = convert_audio(torch.tensor(wav[None,:]), sr, model.sample_rate, model.chin)
```

```python
with torch.no_grad():
    denoised = model(wav[None])[0]
disp.display(disp.Audio(wav.data.cpu().numpy(), rate=model.sample_rate))
disp.display(disp.Audio(denoised.data.cpu().numpy(), rate=model.sample_rate))
```

```python
torchaudio.save('a.wav', denoised, model.sample_rate)
```
